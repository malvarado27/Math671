```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x1,t1: -t1*x1
true_solution=lambda t1: np.exp(-t1**2/2)

tfin=1
dt1=0.02

E=np.zeros(3)

for k in range(0,3):
  ti=0
  xi=1
  dt=dt1/(2**k)
  tn=int(np.ceil(tfin/dt))
  xvec= np.zeros(int(tn)+1)
  t= np.zeros(int(tn)+1)
  xvec[0]=xi
  for i in range(1,tn+1):
    xi=xi+dt*f(xi,ti)
    ti=ti+dt

    xvec[i]=xi
    t[i]=ti

  plt.plot(t,xvec)
  plt.plot(t,true_solution(t))
  plt.show()
  print(xi-true_solution(tfin))
  E[k]=abs(xi-true_solution(tfin))
print(np.log(E[0]/E[1])/np.log(2))
print(np.log(E[1]/E[2])/np.log(2))
```
