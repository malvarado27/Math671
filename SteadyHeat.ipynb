```python
import numpy as np
import matplotlib.pyplot as plt


a=0
b=1

u1=0
u2=1

N=9

h=(b-a)/(N+1)


x=np.linspace(a,b,N+2)
print(h)

print(x)

f = lambda x1 : np.sin(np.pi*x1/2)

true_solution= lambda x1: - (4 / np.pi**2) * np.sin((np.pi * x1) / 2) + (1 + 4 / np.pi**2) * x1

main_diag = -2 * np.ones(N )
off_diag = np.ones(N - 1)
A = np.diag(main_diag) + np.diag(off_diag, k=1) + np.diag(off_diag, k=-1)
A *= 1/ h**2


B = f(x[1:-1])
B[0] -= u1 / h**2
B[-1] -= u2 / h**2

u_internal = np.linalg.solve(A, B)

u_full = np.concatenate(([u1], u_internal, [u2]))


plt.plot(x, u_full, marker='o')
plt.plot(x, true_solution(x))
plt.xlabel('x')
plt.ylabel('Temperature T(x)')
plt.title('1D Steady-State Heat Equation Solution')
plt.grid(True)
plt.show()
```
