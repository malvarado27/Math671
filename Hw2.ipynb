```python
import numpy as np #import the numpy library

y = lambda x : np.cos(x) # define our function
yprime= lambda x : -np.sin(x) # true solution

xo=2 # the x that we are approximating the second derivative at
error=np.zeros(3) #initializing error vector
for i in range(1,4): # iterating 1, 2, 3
  h=0.01/(2**(i-1)) # spatial grid h=0.01/(1) i=1, h=0.01/2 i=2, h=0.01/4 i=3
  yprime_approx=(2*y(xo+h)+3*y(xo)-6*y(xo-h)+y(xo-2*h))/(6*h) # approximating the first derivative
  error[i-1]=abs(yprime(xo)-yprime_approx) #assigning the absolute error to the error vector

print(error) # printing error
print(np.log(error[0]/error[1])/np.log(2)) # printing p using error for h, h/2
print(np.log(error[1]/error[2])/np.log(2)) # printing p using error for h/2, h/4
```
